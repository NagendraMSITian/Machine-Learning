```python
data_groupby=pd.DataFrame(data.groupby(['Customer Name'])[['Quantity','Sales','Profit']].sum().reset_index())
data_groupby
```

```python
feeq_of_customer=pd.DataFrame(data['Customer Name'].value_counts()).reset_index()
```

```python
feeq_of_customer.columns = feeq_of_customer.columns.str.replace('Customer Name','Frequency')
feeq_of_customer.columns = feeq_of_customer.columns.str.replace('index','Customer Name')
feeq_of_customer.columns
```

```python
data_groupby=pd.merge(data_groupby,feeq_of_customer,how='left',on=['Customer Name'])
```

```python
#Total orders from each customer
data_groupby.head()
```

```python
#Average purchase cost on each order for each customer
data_groupby['Avrg_sale']=data_groupby['Sales']/data_groupby['Frequency']
data_groupby.head()
```

```python
#Average profit on each order for each customer
data_groupby['Avrg_Profit']=data_groupby['Profit']/data_groupby['Frequency']
data_groupby.head()
```

```python
#Profit to frequency ratio
data_groupby['Profit_to_Quatity_ratio']=data_groupby['Profit']/data_groupby['Quantity']
data_groupby.head()
```

```python
data_date=data[['Customer Name','Order Date']]
data_date=pd.DataFrame(data_date.groupby(['Customer Name'])['Order Date'].max()).reset_index()
data_date.head()
```

```python
data_groupby=pd.merge(data_groupby,data_date, on='Customer Name')
data_groupby.head()
```

```python
#data_groupby=data_groupby.drop(['Profit_to_Quatity_ratio'],axis=1)
```

```python
data_groupby.columns = data_groupby.columns.str.replace('Order Date','Recent Order')

data_groupby.head()
```

```python
data_groupby['Recent Order']=pd.to_datetime(data_groupby['Recent Order'],yearfirst=True)
data_groupby['Recent Order'].describe()
```

```python
def f(row):
    if pd.to_datetime(row['Recent Order']) > pd.to_datetime(2015-10-30):
        val = 5
    elif pd.to_datetime(row['Recent Order']) <= 2015-10-30 and pd.to_datetime(row['Recent Order']) >pd.to_datetime(2015-8-30):
        val = 4
    elif pd.to_datetime(row['Recent Order']) <= 2015-8-30 and pd.to_datetime(row['Recent Order']) >pd.to_datetime(2015-6-30):
        val = 3
    elif  pd.to_datetime(row['Recent Order']) <= 2015-6-30 and pd.to_datetime(row['Recent Order']) >pd.to_datetime(2015-4-30):
        val = 2
    else:
        val = 1
    return val
data_groupby['Recency_Flag'] = data_groupby.apply(f, axis=1)
#data_groupby = data_groupby.groupby('Customer Name',as_index=False)['Recency_Flag'].max()
```

```python
data_date=data[['Customer Name','Order Date']]
data_date2=pd.DataFrame(data_date.groupby(['Customer Name'])['Order Date'].min().reset_index())
data_date2.head()
```

```python
customer_duration=pd.merge(data_date1,data_date2,on='Customer Name')
```

```python
customer_duration.head()
```

```python
customer_duration['life_time']=customer_duration['Order Date_x']-customer_duration['Order Date_y']
```

```python
customer_duration.head()
```

```python
data_groupby=pd.merge(data_groupby,customer_duration,on='Customer Name')
```

```python
data_groupby.head()
```

```python
col=['Profit_to_Quatity_ratio','Recent Order','Order Date_x','Order Date_y']
data_groupby=data_groupby.drop(col,axis=1)
```

```python
data_groupby.head()
```

```python
X=pd.DataFrame(data_groupby)
X['life_time']=(X['life_time'] / np.timedelta64(1, 'D')).astype(int)
```

```python
X=X.drop(['Customer Name'], axis=1)
```

```python
X['Quantity/life_time']=X['Quantity']/X['life_time']
X=X.drop(['life_time'],axis=1)
X.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
```

```python
kmeans.labels_
```

```python
data_groupby['Segment']=kmeans.labels_
```
